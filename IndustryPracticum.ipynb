```python
import math
for col in df_standardize.columns.values:
    mean=df_standardize[col].mean()
    stdev=df_standardize[col].std()
    #print(col,mean,stdev)
    df_standardize[col]=(df_standardize[col]-mean)/stdev
    if math.isnan(stdev):
        print(col)
        df_standardize=df_standardize.drop(columns=col)

```

```python
from sklearn.cluster import KMeans

#Determining number of clusters
nClusters=range(2,10)
sumDistances=[]
for n in nClusters:
    kmeans=KMeans(n_clusters=n).fit(df_standardize)
    sumDistances.append(kmeans.inertia_) #Proxy for SSE
```

# Year of 2017

```python
basket = (df[df['Dealer_Prov'] =='SK']
          .groupby(['Grower_AIN', 'Prod_Category'])['SumOfSumOfConverted Acres']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Grower_AIN'))
```

```python
basket1 =  (df.groupby(['Dealer_AIN', 'Prod_Category'])['SumOfSumOfConverted Acres']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Dealer_AIN'))
```

```python
basket.head()
```

 - 1 hot encoding

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
#basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
basket_sets=basket_sets.fillna(0)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.02, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[ (rules['lift'] >= 1) &
       (rules['confidence'] >= 0.5) ]
```
